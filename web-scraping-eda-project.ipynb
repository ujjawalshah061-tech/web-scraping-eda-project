```python
#web scraping 
import pandas as pd

url = "https://www.w3schools.com/html/html_tables.asp"
df = pd.read_html(url)[0]

print(df.head())
```

```python
# Basic Data Exploration (EDA)
print("shape of dataset:", df.shape)
```

```python
print(df.info()) #it gives information related dataset
```

```python
print("first 5 row:", df.head()) #first 5 row
```

```python
print("Missing values of check:", df.isnull().sum()) #there is no any null values
```

```python
print("unique countries:", df['Country'].unique()) #there ia total 6 countries
```

```python
print("country count:", df['Country'].value_counts()) #total how many times each country comes
```

```python
print("top 3 countries:", df['Country'].value_counts().head(3)) #its about top 3 countries
```

```python
print("Country percentage distribution:")
print(df['Country'].value_counts(normalize=True) * 100)
```

```python
# Filtering the dataset
print("germany data:", df[df['Country'] == 'Germany'])
```

```python
print(df[(df['Country'] == 'Germany') & (df['Contact'] == 'Maria Anders')])
```

```python
print("sorting data:", df.sort_values('Company'))
```

```python
# country wise company count
print("Country wise company count:", df.groupby('Country').count())
```

```python
#Feature Engineering
df['Company_name_length'] = df['Company'].apply(len)
print("\nDataset with new column:", df)
```

```python
print("\nAverage of Company_name_length") # total average of last column 
print(df['Company_name_length'].mean())
```

## Visualization

```python
# bar chart (Country distribution)
import matplotlib.pyplot as plt
df['Country'].value_counts().plot(kind='bar')
plt.title("Number of companies per country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
df['Country'].value_counts().plot(kind='pie', autopct = '%1.1f%%')
plt.title("Country distribution")
plt.ylabel("")
plt.show()
```

```python
df['Company_name_length'].plot(kind='hist', bins = 5)
plt.title("Company name length distribution")
plt.xlabel("Company name length")
plt.ylabel("frequency")
plt.show()
```

```python
df['Country'].value_counts().sort_values().plot(kind='barh')
plt.xlabel("count")
plt.ylabel("Country")
plt.title("Country-wise companies (sorted)")
plt.show()
```
